```python
def dhoni(x,y):
    y = df['Sixes']
    x = df['Year']
    c = ['yellow','red','blue','purple','pink','green']
    for i in range(len(x)):
        plt.text(x[i],y[i]+0.5, str(y[i]), ha = 'center', color='black')
    plt.bar(x,y,color=c)
    plt.xlabel("Year")
    plt.ylabel("Sixes")
    plt.show()
dhoni(x,y)
    
```
